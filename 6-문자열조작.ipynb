## `counts.most_common(1)[0][0]` 분석

`most_common(n)`에서 `n`은 **"상위 몇 개를 반환할지"** 이고, `[0]`은 그 결과의 인덱스입니다.

---

### 단계별 분해

```python
counts = Counter(['apple', 'banana', 'apple', 'apple', 'banana'])
```

| 표현식 | 결과 | 설명 |
|---|---|---|
| `counts.most_common(1)` | `[('apple', 3)]` | 리스트 (1개짜리) |
| `counts.most_common(1)[0]` | `('apple', 3)` | 리스트의 첫 번째 원소 (튜플) |
| `counts.most_common(1)[0][0]` | `'apple'` | 튜플의 첫 번째 값 **(단어)** |
| `counts.most_common(1)[0][1]` | `3` | 튜플의 두 번째 값 **(개수)** |

---

### 핵심 포인트

- `(1)` → 가장 많은 것 **1개만** 달라는 인자 (인덱스 아님!)
- `[0]` → 1개짜리 리스트에서 첫 번째 원소 꺼내기
- `[0]` → 튜플 `('apple', 3)` 에서 단어만 꺼내기

> `most_common(3)`으로 바꾸면 `[('apple', 3), ('banana', 2), ...]` 처럼 상위 3개가 반환됩니다.

````python
from typing import List
import re
import collections

def mostCommonWord(self, paragraph:str, banned: List[str]) -> str:
    words = [word for word in re.sub(r'[^\w]', ' ', paragraph).lower().split() if word not in banned]

    counts = collections.Counter(words)
    return counts.most_common(1)[0][0] # Return the most common word
````

## `groupAnagrams` — 애너그램 그룹 묶기

애너그램(Anagram)이란 **같은 문자들로 이루어진 단어**를 말합니다.  
예: `"eat"`, `"tea"`, `"ate"` → 모두 `a`, `e`, `t`로 구성 → 같은 그룹

---

### 핵심 아이디어

문자를 **정렬**하면 애너그램끼리 같은 키가 된다는 점을 이용합니다.

```
"eat" → sorted → "aet"
"tea" → sorted → "aet"  ← 같은 키!
"tan" → sorted → "ant"
"nat" → sorted → "ant"  ← 같은 키!
```

---

### 단계별 분해

```python
anagrams = collections.defaultdict(list)
```
> 없는 키에 접근해도 자동으로 빈 리스트 `[]`를 만들어주는 딕셔너리

```python
anagrams[''.join(sorted(word))].append(word)
```

| 단계 | 표현식 | 예시 (`word = "eat"`) |
|---|---|---|
| 1 | `sorted(word)` | `['a', 'e', 't']` |
| 2 | `''.join(...)` | `'aet'` (정렬된 키) |
| 3 | `anagrams['aet'].append('eat')` | `{'aet': ['eat', 'tea', 'ate']}` |

```python
return list(anagrams.values())
```
> 딕셔너리의 값(그룹별 리스트)들을 리스트로 반환

---

### 전체 흐름 예시

```python
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]

# 처리 후 anagrams:
# {
#   'aet': ['eat', 'tea', 'ate'],
#   'ant': ['tan', 'nat'],
#   'abt': ['bat']
# }

# 최종 반환:
# [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]
```

````python
from typing import List

def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
    anagrams = collections.defaultdict(list)

    for word in strs:
        anagrams[''.join(sorted(word))].append(word)

    return list(anagrams.values())
````
